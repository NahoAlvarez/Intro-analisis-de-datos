# Juego de probabilidades:suma de las caras de dos dados ante n eventos

```python
import random

input("¿Quieres tirar los dados? (si/no): ").lower() == 'si'
def tirar_dados():

    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)

    print(f"Tirada de dados: Dado 1 = {dado1}, Dado 2 = {dado2}")
    suma = dado1 + dado2
    print(f"Suma de los dados: {suma}")
    return input("¿Quieres tirar los dados otra vez? (si/no): ").lower() == 'si'

while tirar_dados():
    pass

print("¡Gracias por jugar!")
```
